```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Draw a line from (0,0) to (5, 200)
```

```python
#Make array of coordinates
x = np.array([0,5]) #Save x coordinates
y = np.array([0,200]) #Save y coordinates
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"+")
plt.show()
```

```python
#Muliple points in plot
```

```python
x = np.array([1,2,3,4])
y = np.array([1,2,3,1])
plt.plot(x,y,"*")
plt.show()
```

```python
y = np.array([4,56,43,12,10])
plt.plot(y,'o')
plt.show()
#Taking x axis as 1,2,3,4..
```

```python
y = np.array([4,56,43,12,10])
plt.plot(y,linestyle = 'dotted')
plt.show()
```

```python
y = np.array([4,56,43,12,10])
plt.plot(y,linestyle = 'dashed', color='r')
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title('FIRST PLOT')
plt.show()
```

```python
font1 = {'family':'serif','color':'red','size':'23'}
font2 = {'family':'serif','color':'blue','size':'40'}

y = np.array([4,56,43,12,10])
plt.xlabel('X AXIS', fontdict=font1)
plt.ylabel('Y AXIS', fontdict=font1)
plt.title('FIRST PLOT',fontdict=font2)
plt.show()
```

```python
#fig, ax = plt.subplots()
index = bikes.season.unique()
bar_width = 0.25
df_by_season = bikes.groupby('season')
values = df_by_season.cnt.sum()
plt.bar(index, values, bar_width)
```

```python
d_seasons = df_by_season.cnt.sum().sort_values()
d_seasons
```
