<b>Stacking in NumPy

Stacking: Several arrays can be stacked together along different axes.<br>
np.vstack: To stack arrays along vertical axis.<br>
np.hstack: To stack arrays along horizontal axis.<br>
np.column_stack: To stack 1-D arrays as columns into 2-D arrays.<br>
np.concatenate: To stack arrays along specified axis (axis is passed as argument).<br>

```python
a = np.array([[1,2,3],[3,2,1]])
b = np.array([[4,5,6],[6,5,4]])

#vsatck
print(np.vstack((a,b)),"\n")

#hstack(
print(np.hstack((a,b)),"\n")

#column stack
c = np.array([5000, 6000])
print(np.column_stack((a, c)),"\n")

#concatenate
print(np.concatenate((a,b),axis=0), "\n")
print(np.concatenate((a,b),axis=1))
```

<b>Splitting:For splitting, we have these functions: <br></b>
np.hsplit: Split array along horizontal axis.<br>
np.vsplit: Split array along vertical axis.<br>
np.array_split: Split array along specified axis.<br>

```python
arr = np.array([[1,2,3,4],[4,5,6,7],[5,6,7,8],[7,8,9,0]])
print(arr,"\n")
print(np.hsplit(arr,2),"\n")
print(np.vsplit(arr,2),"\n")
print(np.array_split(arr,2,axis=0),"\n")
```
